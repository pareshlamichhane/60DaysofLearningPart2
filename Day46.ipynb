```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python

print(a + b)        # Vector addition
print(a * 2)        # Scalar multiplication
print(a.shape)      # Shape of array
```

```python
# Broadcasting example
A = np.array([[1,2,3],[4,5,6]])
B = np.array([1,2,3])
print(A + B)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'value': np.random.randn(1000),
    'group': np.random.choice(['A', 'B'], 1000)
})

sns.histplot(data=df, x='value', hue='group', kde=True)
```

```python
# Broadcasting
x = np.array([[1], [2], [3]])
y = np.array([10, 20, 30])
print(x + y)
```

```python
import pandas as pd

data = {
    'name': ['Ram', 'Shyam', 'Hari'],
    'age': [25, 30, 35]
}

df = pd.DataFrame(data)
print(df)

# Filter
print(df[df['age'] > 28])
```

```python
import seaborn as sns
import numpy as np
import pandas as pd

# Random dataset
df = pd.DataFrame({
    'group': np.random.choice(['A', 'B'], 100),
    'value': np.random.randn(100)
})

sns.histplot(data=df, x='value', hue='group', kde=True)
```
