```python
import numpy as np
```

```python
# Set seed for reproducibility
np.random.seed(42)
```

```python
# Simulate temperature data (7 days, 24 hours)
temps = np.random.normal(loc=25, scale=5, size=(7, 24))  # mean 25째C, std dev 5
```

```python
# Example: print Monday's temps
print("Monday Temps:", temps[0])
```

```python
# Average temp per day
daily_avg = temps.mean(axis=1)
print("Daily Average Temps:", daily_avg)
```

```python
# Max temp overall
print("Hottest Hour:", temps.max())
```

```python
# Which day was hottest on average?
print("Day with highest avg temp:", np.argmax(daily_avg))
```

```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

```python
# Line plot of daily average temperatures
plt.plot(days, daily_avg, marker='o', linestyle='-', color='tomato')
plt.title("Average Daily Temperatures")
plt.xlabel("Day")
plt.ylabel("Temperature (째C)")
plt.grid(True)
plt.show()
```

```python
hot_hours = temps > 30
print("Number of hot hours (>30째C):", np.sum(hot_hours))
```

```python
mean_temp = temps.mean()
std_temp = temps.std()
z_scores = (temps - mean_temp) / std_temp
print("Standardized temps (z-scores):", z_scores)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(temps, index=days)
df.to_csv("weekly_temps.csv", index_label="Day")
```
